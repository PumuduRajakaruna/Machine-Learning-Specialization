```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x_train = np.array([1.0, 2.0])           #(size in 1000 square feet)
y_train = np.array([300.0, 500.0])           #(price in 1000s of dollars)

```

The code below calculates cost by looping over each example. In each loop:

f_wb, a prediction is calculated 

the difference between the target and the prediction is calculated and squared.

this is added to the total cost.


```python
def compute_cost(x, y, w, b):
    
    # number of training examples
    m = x.shape[0]
    
    cost_sum = 0
    for i in range(m):
        f_wb = w*x[i] + b
        cost = (f_wb - y[i]) ** 2 
        cost_sum = cost_sum + cost
    total_cost = (1/(2*m)) * cost_sum

    return total_cost 

# assume 
b = 100
total_cost = []
w_values = np.linspace(-250, 250, 500)
for i in range(len(w_values)):
    total_cost.append(compute_cost(x_train, y_train, i, b))

plt.plot(w_values,total_cost)
plt.xlabel('w values')
plt.ylabel('Total Cost')
plt.title('Total Cost vs. w values')
plt.show()
```

## How the cost function value varies with the w and b values 

By this, we can find suitable w and b values such that cost function will get a minimum value

```python
x_train = np.array([1.0, 1.7, 2.0, 2.5, 3.0, 3.2])
y_train = np.array([250, 300, 480,  430,   630, 730,])

b_values = np.linspace(-250, 250, 500)
w_values = np.linspace(-250, 250, 500)
z = np.zeros((len(w_values), len(b_values)))

for i in range(len(w_values)):
    for j in range(len(b_values)):
        z[i, j] = compute_cost(x_train, y_train, w_values[i], b_values[j])

x, y = np.meshgrid(w_values, b_values)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z)

# Set labels and title
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('3D Plot')

# Show the plot
plt.show()
```
